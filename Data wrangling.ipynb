```python
import pandas as pd
```

```python
df_cases = pd.read_csv("./CoronavirusPL - Timeseries.csv")
df_terc = pd.read_csv("./TERC_Urzedowy_2020-03-30.csv", delimiter=";")
df_simc = pd.read_csv("./SIMC_Urzedowy_2020-03-30.csv", delimiter=";")
df_cases[-10:]
```

```python
df_pow = df_terc.loc[(df_terc.NAZWA_DOD == "powiat") | (df_terc.NAZWA_DOD == "miasto na prawach powiatu"), ["WOJ", "POW", "NAZWA"]]
df_pow.rename(columns={"NAZWA": "NAZWA_POW"}, inplace=True)
```

```python
province_city_loc = pd.DataFrame([
    [22, 54.366667, 18.633333],
    [2, 51.1078852, 17.0385376],
    [4, 53.0137500, 18.5981400],
    [6, 51.246452, 22.568445],
    [8, 51.935619, 15.506186],
    [10, 51.776667, 19.454722],
    [12, 50.061389, 19.938333],
    [14, 52.232222, 21.008333],
    [16, 50.664722, 17.926944],
    [20, 53.135278, 23.145556],
    [18, 50.033611, 22.004722],
    [24, 50.264167, 19.023611],
    [26, 50.874167, 20.633333],
    [28, 53.773056, 20.476111],
    [30, 52.408333, 16.934167],
    [32, 53.438056, 14.542222]
    ], columns=["WOJ", "lat", "long"])
```

```python
df_woj = df_terc.loc[(df_terc.NAZWA_DOD == "województwo"), ["WOJ", "NAZWA"]]
df_woj.rename(columns={"NAZWA": "NAZWA_WOJ"}, inplace=True)
df_woj = pd.merge(df_woj, province_city_loc, on="WOJ", how="right")
df_woj
```

```python
city_county = pd.merge(
    df_simc, 
    pd.merge(df_pow, df_woj, how="right", on="WOJ"), 
    how="right", on=["WOJ", "POW"])[["NAZWA", "NAZWA_POW", "NAZWA_WOJ", "lat", "long"]]
city_county.rename(columns={"NAZWA": "City", "NAZWA_POW": "County", "NAZWA_WOJ": "Province"}, inplace=True)
len(city_county), len(df_simc)
```

```python
df_cases['Province'] = df_cases['Province'].apply(lambda s: s.upper())
pd.merge(df_cases, city_county, how="inner", on=["City", "Province"])
```

```python
df_cases.loc[df_cases.City.isnull()]
```

```python
new_cases = df_cases.groupby(["Timestamp","Province"]).size()
new_cases.index = new_cases.index.set_levels(pd.DatetimeIndex(new_cases.index.levels[0], dayfirst=True), level=0)
new_cases = new_cases.unstack(1).resample('1D').sum().stack().swaplevel()
new_cases
```

```python
total_cases = new_cases.copy()
for province in total_cases.index.levels[0]:
    total_cases[province] = new_cases[province].cumsum()
total_cases
```

```python
df = pd.concat({"new": new_cases, "total": total_cases}, axis=1)
for date, new_df in df.groupby(level=1):
    new_df = new_df.merge(df_woj.rename(columns={"NAZWA_WOJ": "Province"}), on="Province", how="right")
    new_df.to_csv("./data-daily-woj/covid19-poland-woj-{:%Y-%m-%d}.csv".format(date))
```
