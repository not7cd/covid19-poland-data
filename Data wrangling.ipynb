```python
import pandas as pd
```

```python
df_cases = pd.read_csv("./CoronavirusPL - Timeseries.csv")
df_terc = pd.read_csv("./TERC_Urzedowy_2020-03-30.csv", delimiter=";")
df_simc = pd.read_csv("./SIMC_Urzedowy_2020-03-30.csv", delimiter=";")
```

```python
df_woj = df_terc.loc[(df_terc.NAZWA_DOD == "województwo"), ["WOJ", "NAZWA"]]
df_woj.rename(columns={"NAZWA": "NAZWA_WOJ"}, inplace=True)
```
